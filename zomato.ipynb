```python
#Python Project On Top 1000 Bollywood Movies And Their Box Office
import seaborn as sns
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
zom=pd.read_csv("Zomatodata.csv")
zom
```

```python
#Data Cleaning And Data Processing
#1 to modify the rate by only printing rating that 4.1,3,7 etc removing "/" part [Converting The Data Type Of Rate Column]
def rating(value):
    value=str(value).split("/")
    value=value[0];
    return float(value)

zom["rate"]=zom["rate"].apply(rating)
zom
```

```python
#All Columns Are False Means No Null Values
```

```python
#Q1 Types Of Resturant Which types is used by cosumer more?
#We Can Show It By Bar Graph Seaborn Library

plt.title("Types Of Resturants Bar Plot")
sns.countplot(x=zom["listed_in(type)"],hue=zom["listed_in(type)"],palette="Set2")
plt.show()
```

```python
#Conclusion-Customer prefer the dining type in resturant
```

```python
#Q2 How many votes has each type of resturant recieved from customer
#We can show it by line plot/graph "x=types of resturant","y=Votes"

plt.title("Votes As Per Resturant Type")
group=zom.groupby("listed_in(type)")["votes"].sum()
result=pd.DataFrame({"votes":group})
plt.xlabel("Types Of Resturant")
plt.ylabel("Votes")
plt.plot(result,c="y",marker="o")
plt.show()
```

```python
#Conclusion-Dining Has Got More Votes From All Of Other Types
```

```python
#Q3 What are the rating majority of resturant recieved by the customer
#We Will Show It By Histogram

plt.title("Rating Majority By Histogram")
plt.hist(zom["rate"],bins=5,color="skyblue")
plt.show()
```

```python
#Conclusion-The Higest Rating Is From 3.50 To 4.25
```

```python
#Q4 What is the average spending money on couple orders
sns.countplot(x=zom["approx_cost(for two people)"],hue=zom["approx_cost(for two people)"],palette="Set2")
plt.show()
```

```python
#Conclusion-Here Couple Spends 300 Rupees To Order Food Online 
```

```python
#Q5-Which Mode has recieved the maximum rating online or offline
#we can see it using boxplot
plt.figure(figsize=(6,6))
plt.title("Box Plot Online Order")
sns.boxplot(x="online_order",y="rate", hue="online_order",data=zom,palette="Set2")
plt.show()
```

```python
#Conclusion-As per box plot we can say that online order has maximum rating
```

```python
#Q6-Which restaurant recieved more offline order so that zomato can provide customer good offers
pivot_table=zom.pivot_table(index="listed_in(type)",columns="online_order",aggfunc="size",fill_value=0)
sns.heatmap(pivot_table,annot=True,cmap="YlGnBu",fmt="d")
plt.show()
```

```python
#Conclusion:Dinning Mainly Accept Offline Orders Whereas Cafe Primarily Recieve Online Orders.This Tells That Clients Prefers Orders In Person At Restaurants,But Prefer Online Order At Cafes!!!
```
